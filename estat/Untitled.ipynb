```python
import numpy as np
import matplotlib.pyplot as plt

# Dados fornecidos
x = np.array([0, 0, 0, 15, 15, 15, 30, 30, 30, 45, 45, 45, 60, 60, 60, 75, 75, 75])
y = np.array([8, 6, 8, 12, 10, 14, 25, 21, 24, 31, 33, 28, 44, 39, 42, 48, 51, 44])

# 1. Calcular as médias de x e y
x_mean = np.mean(x)
y_mean = np.mean(y)

# 2. Calcular (x - x_mean) * (y - y_mean) e (x - x_mean)^2
numerator = np.sum((x - x_mean) * (y - y_mean))
denominator = np.sum((x - x_mean) ** 2)

# 3. Coeficiente angular (b)
b = numerator / denominator

# 4. Coeficiente linear (a)
a = y_mean - b * x_mean

# 5. Estimativa para x = 50
y_estimate = a + b * 50

# Resultados finais
print("Equação da reta de regressão: y = {:.4f} + {:.4f}x".format(a, b))
print("Estimativa para x = 50: y = {:.3f}".format(y_estimate))

# 6. Criar o gráfico
plt.scatter(x, y, color='blue', label='Dados observados')  # Pontos de dispersão
plt.plot(x, a + b * x, color='red', label='Reta de regressão')  # Reta de regressão
plt.title('Reta de Regressão Linear')
plt.xlabel('Temperatura (°C)')
plt.ylabel('Quantidade (gramas)')
plt.legend()
plt.grid(True)
plt.show()
```

 - Coeficiente angular = 0.5676
 - Coeficiente linear = 5.8254
 - Como coeficiente angular é maior que 0 isso indica que a variavel dependente (quantidade) cresce a medida que a variavel independente temperatura tambem cresce
 - Com o grafico é possivel analisar que a quantidade prevista para 50⁰ é 34.206
